```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


```

```python
# Combine the data into a single dataset
data_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')

# Display the data table for preview
data_df.head()
```

```python
# Checking the number of mice - this ended up just being the amount of times a mouse was given a medication
#Trying again in next box
number_mice = data_df["Mouse ID"].value_counts()
number_mice
```

```python
#Take two finding the number of mice
ColumnID = data_df.loc[:, ["Mouse ID"]]
ColumnID.drop_duplicates().count()

#There are 249 Mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

dup_mice = data_df.loc[data_df.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()

dup_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
drop_dup=data_df.loc[data_df["Mouse ID"].isin(dup_mice)==False]
drop_dup
```

```python
# Checking the number of mice in the clean DataFrame.
drop_dup.count()
unique_mice=drop_dup["Mouse ID"].nunique()
unique_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

# Group data by Drug Regimen
summary_df = data_df.groupby('Drug Regimen')

# Calculate the mean, median, standard deviation, and sem
tumor_mean = summary_df['Tumor Volume (mm3)'].mean()
tumor_median = summary_df['Tumor Volume (mm3)'].median()
tumor_stdev = summary_df['Tumor Volume (mm3)'].std()
tumor_sem = summary_df['Tumor Volume (mm3)'].sem()

# Create DataFrame
summary_grouped_df = pd.DataFrame({'Mean': tumor_mean, 'Median': tumor_median,'Standard Deviation': tumor_stdev, 'SEM': tumor_sem})

summary_grouped_df.head()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
capomulin = drop_dup.loc[(drop_dup["Drug Regimen"] == "Capomulin"),:]
capomulin
# pick a mouse
cap_mouse = capomulin.loc[(capomulin["Mouse ID"] == "x401"),:]
cap_mouse

# Set x and y axis
x = cap_mouse["Timepoint"]
y = cap_mouse["Tumor Volume (mm3)"]

# Plot
time_volume_line, = plt.plot(x, y)

# labels
plt.title('Tumor volume over time for Mouse x401 treated with Capomulin')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume(mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

mouse_weight = Capomulin.groupby(Capomulin["Mouse ID"])["Weight (g)"].mean()
avg_volume = Capomulin.groupby(Capomulin["Mouse ID"])["Tumor Volume (mm3)"].mean()

#Scatter Plot 
plt.scatter(mouse_weight, avg_volume)
plt.title("Weight versus Tumor volume for the Capomulin regimen")
plt.xlabel("Avg. Weight of Mouse")
plt.ylabel("Avg. Tumor Volume")
plt.show()
```
